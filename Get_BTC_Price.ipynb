```python
import gdax
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#auth
public_client=gdax.PublicClient()
```

```python
#Collects the data of Bitcoin over an hour intervals
time_data=public_client.get_product_historic_rates('BTC-USD', start='2018-04-12T16:42:48Z', end='2018-04-14T16:42:48Z', granularity=3600)
```

```python
#Inputs the data into a Data Frame
df=pd.DataFrame(time_data, columns=["Time", "low","high", "open","close", "vol"])
```

```python
#Averages the "open" price and the "close" price.
avg_btc=[]
avg_btc=(df['open']+df['close'])/2
```

```python
"Creates the DataFrame for the average prices"
pricedf=pd.DataFrame(avg_btc, columns=["Average Price"])
```

```python
#some Extra features of the data
plt.plot(avg_btc)
plt.ylabel("Bitcoin Price")
plt.xlabel("Hours")
plt.title("Bitcoin Price over 2 days")

stats.describe(avg_btc)
```
